```python
import pandas as pd
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://root:admin@localhost:5432/ny_taxi')
```

```python
engine.connect()
```

```python
query = """
SELECT 1 as number;
"""

pd.read_sql(query, con=engine)
```

```python
query = """
SELECT *
FROM pg_catalog.pg_tables
WHERE schemaname != 'pg_catalog' AND 
    schemaname != 'information_schema';
"""

pd.read_sql(query, con=engine)
```

#### Adding trip data
https://www.nyc.gov/assets/tlc/downloads/pdf/data_dictionary_trip_records_green.pdf

```python
df = pd.read_csv('green_tripdata_2019-01.csv')
df.head(10)
```

```python
print(pd.io.sql.get_schema(df, name='green_tripdata'))
```

```python
df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
```

```python
print(pd.io.sql.get_schema(df, name='green_tripdata', con=engine))
```

```python
df.to_sql(name='green_tripdata', con=engine, index=False)
```

```python
query = """
SELECT * FROM green_tripdata LIMIT 10
"""

pd.read_sql(query, con=engine)
```
